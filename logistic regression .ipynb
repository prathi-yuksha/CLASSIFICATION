```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("clean_social.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l = LabelEncoder()
df["productivity"] = l.fit_transform(df["productivity"])
df
```

```python
l = LabelEncoder()
df["gender"] = l.fit_transform(df["gender"])
df
```

```python
l = LabelEncoder()
df["useSocial"] = l.fit_transform(df["useSocial"])
df
```

```python
l = LabelEncoder()
df["typeSocial"] = l.fit_transform(df["typeSocial"])
df
```

```python
df.drop(columns=["education","profession","workDuration"],axis=1)
```

```python
x = df[["useSocial","socialDuration","typeSocial"]]
y = df["productivity"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.9,random_state=0)
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
pred
```

```python
from sklearn.metrics import  accuracy_score
accuracy_score(pred,y_test)
```
